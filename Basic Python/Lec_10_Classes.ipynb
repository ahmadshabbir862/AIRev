# Classes in Python

## Writing a Class
Python uses a keyword "class" to define a Class

    #body of class

## What Class Actually holds

* A class may hold variables
* A class may hold functions

+ Example
* A class car has color, model and seating capacity are 'attributes' of class
* A car can run, stop, speed-up, speed-down are 'behaviours' of class car

## Creating an instance of class

class Car():
    * Define body of class
    * Define attributes and behaviours
    
Creating objects/instances of Car Class
* car1 = Car()
* car2 = Car()

```python
#define a class
class Car():
    # attributes >>> are variables
    # behaviours >>> functions
    
    #Attributes
    def __init__(self, model, price, color,year): # first parameter will always be self
        self.model = model
        self.price = price
        self.color = color
        self.year = year
        self.battery = "200amp" #default attribute
        
    # Behaviours
    def descriptionCar(self):
        print(f"\nThe Model of the car {self.model}")
        print(f"The Price of the  Car is: {self.price}")
        print(f"The Color of the Car is : {self.color}")
        print(f"The Year of the car manufacturing : {self.year}")
        
    def move(self):
        print(f"\n {self.model} is moving with speed")
        
    def applyingBreaks(self):
        print(f"\n {self.model} has applied breaks")
        
# Object / Instance of a Class

car1 = Car("AAA","10K", "Orange", 2015)
car1.descriptionCar()
car1.move()
car1.applyingBreaks()



car2 = Car("BBB", "15K", "blue", 2018)
car2.descriptionCar()
car2.move()
car2.applyingBreaks()
```
